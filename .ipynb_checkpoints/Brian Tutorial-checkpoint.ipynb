```python
from brian2 import *
import numpy as np
%matplotlib inline
```

```python
start_scope()

N = 2
tau = 10*ms
eqs = '''
dv/dt = (1-v)/tau : 1
'''
# G = NeuronGroup(1, eqs)
G = NeuronGroup(N, eqs,threshold='v>0.4', reset='v = 0', method='euler')
G.v = 'rand()'

# record = which cell we are recording
M = StateMonitor(G, 'v', record=True)
spikemon = SpikeMonitor(G)
run(10*ms)
```

```python
plot(M.t/ms, M.v[0])
xlabel('Time (ms)')
ylabel('v');
print('Spike times: %s' % spikemon.t[:])
```

```python
plot(spikemon.t/ms, spikemon.i, '.k')
xlabel('Time (ms)')
ylabel('Neuron index');
```

```python
plot(M.t/ms, 1-exp(-M.t/tau), 'C1--',label='Analytic')
```
